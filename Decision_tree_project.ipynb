```python
#model building
x = data.drop('quality', axis=1)
y = data['quality']
```

```python
x.head()
```

```python
y.head()
```

```python
#split x and y into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state=0)
```

```python
#check the shape of x_train,x_test
x_train.shape,x_test.shape
```

```python
#check datatypes in x_test
x_test.dtypes
```

```python
#import decision tree classifier
from sklearn.tree import DecisionTreeClassifier
#instantiate the decisiontree classifier model with criterion gini index
DTree= DecisionTreeClassifier(criterion= 'gini',max_depth= 3,random_state= 0)
```

```python
#fit the model
DTree.fit(x_train,y_train)
```

```python
y_pred= DTree.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
#evalute the model
print('model accuracy score with criterion index:{0:0.5f}'. format(accuracy_score(y_test,y_pred)))
```

```python
#create confusion matrix
from sklearn.metrics import confusion_matrix
conf_matrix= confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
#display the confusion matrix using seaborn heatmap
plt.figure(figsize=(10,6))
sns.heatmap(conf_matrix,annot= True, fmt= 'd',cmap= 'crest', cbar= False)
```

```python
#create classification report
from sklearn.metrics import classification_report
class_report= classification_report(y_test,y_pred)
print(class_report)
```

```python
plt.figure(figsize=( 12, 6))
from sklearn import tree
tree.plot_tree(DTree.fit(x_train,y_train))
```
